```python
data_val = data[:100].copy()
data = data.drop(data_val.index)
```

```python
x = data.drop('Rating', axis=1)
y = data['Rating']
data_val_rating = data_val['Rating']
data_val.drop('Rating', axis=1, inplace=True)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.15,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

models = []
models.append(['XGBRegressor', XGBRegressor()])
models.append(['DecisionTree', DecisionTreeRegressor(random_state=0)])
models.append(['RandomForest', RandomForestRegressor(random_state=0)])
```

```python
data
```

```python
grid_models = [
  
    (DecisionTreeRegressor(), [{'criterion': ['poisson'], 'random_state': [0]}]),
    (RandomForestRegressor(), [{'n_estimators': [150, 200, 250], 'criterion': ['squared_error', 'friedman_mse', 'poisson'], 'random_state': [0]}]),
    (XGBRegressor(), [{'n_estimators': [150, 200, 250], 'learning_rate': [0.1, 0.01, 0.001], 'max_depth': [3, 5, 7]}]),
]
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns


best_r2_score = -float('inf')
best_params = {}
best_model = None

for model, param_grid in grid_models:
    grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=10)
    grid.fit(xtrain, ytrain)
    
    neg_mse = grid.best_score_
    r2 = r2_score(data_val_rating, grid.predict(data_val))
    params = grid.best_params_
    
    print('{}: \n Best Negative MSE: {:.2f}'.format(model.__class__.__name__, neg_mse))
    print('Best R-squared: {:.2f}'.format(r2))
    print('Best Params: {}'.format(params))
    print('')
    print('-' * 25)
    print('')
    
    if r2 > best_r2_score:
        best_r2_score = r2
        best_params = params
        best_model = model

# Create the best model with the best hyperparameters
best_model = best_model.set_params(**best_params)

# Train the best model on the training data
best_model.fit(xtrain, ytrain)

# Make predictions on the test data
y_pred = best_model.predict(data_val)

# Calculate evaluation metrics for regression
mse = mean_squared_error(data_val_rating, y_pred)
r2 = r2_score(data_val_rating, y_pred)

t_pred = best_model.predict(xtest)
t_mse = mean_squared_error(ytest, t_pred)
t_r2 = r2_score(ytest, t_pred)

# Print evaluation metrics
print("-----------------")
print("On Test Data:")
print("Mean Squared Error (MSE):", t_mse)
print("R-squared:", t_r2)
print("-----------------")
print("On Validation Data:")
print("Best R-squared:", best_r2_score)
print("Mean Squared Error (MSE):", mse)
```
