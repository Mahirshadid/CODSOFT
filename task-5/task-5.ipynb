```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data['Amount_scaled'] = scaler.fit_transform(data[['Amount']])
```

```python
data
```

```python
data.drop('Amount',axis=1,inplace=True)
data.drop('Time',axis=1,inplace=True)
x = data.drop('Class',axis=1)
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state = 2)
xtrain_smote, ytrain_smote = smote.fit_resample(xtrain, ytrain.ravel())
  
print('Shape of xtrain: ', xtrain_smote.shape)
print('Shape of xtrain: ', ytrain_smote.shape)
```

```python
xtrain_smote
```

```python
ytrain_smote
```

```python
plt.hist(ytrain_smote)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

rf = RandomForestClassifier(n_estimators=50)

rf.fit(xtrain_smote, ytrain_smote)

y_pred = rf.predict(xtest)

acc = accuracy_score(ytest, y_pred)
precision = precision_score(ytest, y_pred, average='weighted', zero_division=0)
recall = recall_score(ytest, y_pred, average='weighted', zero_division=0)
f1 = f1_score(ytest, y_pred, average='weighted', zero_division=0)

print("Accuracy:", acc)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
